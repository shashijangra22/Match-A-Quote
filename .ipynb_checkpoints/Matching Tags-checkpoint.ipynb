```python
import numpy as np
import pandas as pd
```

```python
iDataSet=pd.read_csv("imagedataset.csv")
qDataSet=pd.read_csv("Filtered_Quotes(0-1500).csv")
```

```python
images=np.array(iDataSet)
quotes=np.array(qDataSet)
```

```python
def MatchAQuote(qCats,quote,qTags):
    tempScore=[]
    for image in images:
        iName=image[0]
        iCats=image[1]
        iTags=image[2]
        score1=matchWords(qCats.split(','),iCats.split(','))
        score2=matchWords(qTags.split(','),iTags.split(','))
        if score1 and score2:
            tempScore.append((iName,score1,score2))
    return tempScore    
```

```python
def matchWords(a,b):
    score=0
    for w in a:
        for i in b:
            if w.lower()==i.lower():
                score+=1
    return score
```

```python
scores=[]
for sample in quotes:
    cats=sample[0]
    quote=sample[1]
    tags=sample[2]
    tempScore=MatchAQuote(cats,quote,tags)
    scores.append(tempScore)
```
